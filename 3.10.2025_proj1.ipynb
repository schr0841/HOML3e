```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Generate random data
np.random.seed(0)
X1 = np.random.rand(50) * 100
Y1 = np.random.rand(50) * 100
Z1 = 2.5 * X1 + 1.5 * Y1 + np.random.rand(50) * 50

X2 = np.random.rand(50) * 100
Y2 = np.random.rand(50) * 100
Z2 = 3.0 * X2 + 2.0 * Y2 + np.random.rand(50) * 50

# Fit linear regression for both sets
reg1 = LinearRegression().fit(np.column_stack((X1, Y1)), Z1)
reg2 = LinearRegression().fit(np.column_stack((X2, Y2)), Z2)

# Create a meshgrid for the plane
xx1, yy1 = np.meshgrid(np.linspace(X1.min(), X1.max(), 10), np.linspace(Y1.min(), Y1.max(), 10))
zz1 = reg1.predict(np.column_stack((xx1.ravel(), yy1.ravel()))).reshape(xx1.shape)

xx2, yy2 = np.meshgrid(np.linspace(X2.min(), X2.max(), 10), np.linspace(Y2.min(), Y2.max(), 10))
zz2 = reg2.predict(np.column_stack((xx2.ravel(), yy2.ravel()))).reshape(xx2.shape)

# Create plots
fig = plt.figure(figsize=(12, 6))

# Left plot
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X1, Y1, Z1, c='k')
ax1.plot_surface(xx1, yy1, zz1, color='c', alpha=0.5)
ax1.set_xlabel('Gew')
ax1.set_ylabel('MsA')
ax1.set_zlabel('Bein/R')

# Right plot
ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(X2, Y2, Z2, c='k')
ax2.plot_surface(xx2, yy2, zz2, color='c', alpha=0.5)
ax2.set_xlabel('Gew')
ax2.set_ylabel('MsNa')
ax2.set_zlabel('Bein/R')

plt.show()
```

## Python Code to Generate Similar Plots
Here’s a Python code snippet using matplotlib and numpy to create a close approximation of the images:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Generate sample data
np.random.seed(0)
Gew = np.random.rand(50) * 100
BeinIR = np.random.rand(50) * 30 + 85
MsA = 0.5 * Gew + 0.3 * BeinIR + np.random.rand(50) * 10
MsNa = 0.4 * Gew + 0.4 * BeinIR + np.random.rand(50) * 10

# Create 3D scatter plot for MsA
fig = plt.figure(figsize=(14, 6))

# Left Plot - MsA
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(Gew, BeinIR, MsA, c='black', s=50)
ax1.set_xlabel('Gew')
ax1.set_ylabel('BeinIR')
ax1.set_zlabel('MsA')

# Fit a plane for MsA
X1 = np.vstack((Gew, BeinIR)).T
model1 = LinearRegression().fit(X1, MsA)
xx1, xx2 = np.meshgrid(np.linspace(0, 100, 10), np.linspace(85, 115, 10))
zz1 = model1.intercept_ + model1.coef_[0] * xx1 + model1.coef_[1] * xx2
ax1.plot_surface(xx1, xx2, zz1, alpha=0.5, cmap='viridis')

# Right Plot - MsNa
ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(Gew, BeinIR, MsNa, c='black', s=50)
ax2.set_xlabel('Gew')
ax2.set_ylabel('BeinIR')
ax2.set_zlabel('MsNa')

# Fit a plane for MsNa
X2 = np.vstack((Gew, BeinIR)).T
model2 = LinearRegression().fit(X2, MsNa)
zz2 = model2.intercept_ + model2.coef_[0] * xx1 + model2.coef_[1] * xx2
ax2.plot_surface(xx1, xx2, zz2, alpha=0.5, cmap='viridis')

plt.show()
```
